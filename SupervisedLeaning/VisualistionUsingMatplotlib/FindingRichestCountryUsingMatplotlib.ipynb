#### Notice that ‘UAE’ has the highest average gdp_percap. Having this information, let’s look at ‘UAE’ in more details to find out if it’s actually the most richest country in the world on a per-person basis.

As its not clear from above picture, lets draw a barplot

```python
UAE.plot.bar(x='year',y='gdp_percap')
plt.xlabel("Year")
plt.ylabel("GDP/Capita")
```

```python
plt.subplot(3,1,1)
plt.title('GDP Per Capita')
plt.plot(UAE['year'],UAE['gdp_percap'])
plt.subplot(3,1,2)
plt.title('GDP in Billions')
plt.plot(UAE['year'],UAE['population']*UAE['gdp_percap']/10**9)
plt.subplot(3,1,3)
plt.title('Population in millions')
plt.plot(UAE['year'],UAE['population']/10**6)
plt.tight_layout()
```

**Similarly for a clearer result, let’s also plot it’s bar graph:**



```python
plt.subplot(3,1,1)
plt.title('GDP Per Capita')
plt.bar(UAE['year'],UAE['gdp_percap'])
plt.subplot(3,1,2)
plt.title('GDP in Billions')
plt.bar(UAE['year'],UAE['population']*UAE['gdp_percap']/10**9)
plt.subplot(3,1,3)
plt.title('Population in millions')
plt.bar(UAE['year'],UAE['population']/10**6)
#plt.tight_layout()
```

As clear from plot and bar graph with increase in GDP, there is increase in population too and increase in GDP growth rate is less as compared to increase in population. So, there is overall decrease in gdp per capita from 2007 to 2015. 

```python
plt.plot(UAE['year'], (UAE['population']/UAE['population'].iloc[0]*100))
uae_gdp = UAE['population'] * UAE['gdp_percap']
plt.plot(UAE['year'], uae_gdp/uae_gdp.iloc[0]/100)
plt.plot(UAE['year'], UAE['gdp_percap']/UAE['gdp_percap'].iloc[0]/100)
plt.title('Gdp and Population growth in Dubai(first year = 100)')
plt.legend(['Population', 'GDP', 'GDP Per Capita'], loc=2)
```
